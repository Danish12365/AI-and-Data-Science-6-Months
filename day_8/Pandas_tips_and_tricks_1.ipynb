# 9-convert string to number

```python
df = pd.DataFrame({'A': [1, 2, 3, 4, 5, 6, 7, 8], 'B': [4, 5, 6, 7, 8, 9, 10, 11]})
df
```

```python
df.dtypes
```

```python
df.astype({'A': 'object', 'B': 'object'}).dtypes
```

Functionality:

Selects Column: The code first extracts the data from column 'A' of the DataFrame df.\
Attempts Conversion: It then tries to convert each value in that column to a numeric data type (e.g., integer or float).\
Handles Errors (coerce): If any value cannot be converted (e.g., because it's a string like "apple"), the errors='coerce' argument ensures that those values are replaced with NaN instead of raising an error.

```python
pd.to_numeric(df['A'], errors='coerce')
df.dtypes
```

# 16-spliting a string into multiple columns

```python
df = pd.DataFrame({'Name':['Asiif Khan','Naveed Ahmed','Kashif Khan','Asiif Khan','Muhammad Aammar','Whaeed Ahmad'],
                  'Age':['20','21','20','20','21','20'],
                  'City':['Karachi pakistan','Multan pakistan','Fasilbad pakistan','Bahawalpur pakistan','Larkana pakistan','Mari pakistan']})
df
```

```python
df.Name.str.split(' ',expand=True)
```

```python
df[['First Name','Last Name']] = df.Name.str.split(' ',expand=True)
df
```

```python
df.City.str.split('  ',expand=True)
```

```python
df[['City','Country']] = df.City.str.split('  ',expand=True)
df
```
