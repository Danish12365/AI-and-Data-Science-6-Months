# 9-convert string to number

```python
df = pd.DataFrame({'A': [1, 2, 3, 4, 5, 6, 7, 8], 'B': [4, 5, 6, 7, 8, 9, 10, 11]})
df
```

```python
df.dtypes
```

```python
df.astype({'A': 'object', 'B': 'object'}).dtypes
```

Functionality:

Selects Column: The code first extracts the data from column 'A' of the DataFrame df.\
Attempts Conversion: It then tries to convert each value in that column to a numeric data type (e.g., integer or float).\
Handles Errors (coerce): If any value cannot be converted (e.g., because it's a string like "apple"), the errors='coerce' argument ensures that those values are replaced with NaN instead of raising an error.

```python
pd.to_numeric(df['A'], errors='coerce')
df.dtypes
```

# 16-spliting a string into multiple columns

```python
# Create DataFrame
df = pd.DataFrame({
    'Name': ['Asiif Khan', 'Naveed Ahmed', 'Kashif Khan', 'Asiif Khan', 'Muhammad Aammar', 'Whaeed Ahmad'],
    'Age': ['20', '21', '20', '20', '21', '20'],
    'City': ['Karachi pakistan', 'Multan pakistan', 'Fasilbad pakistan', 'Bahawalpur pakistan', 'Larkana pakistan', 'Mari pakistan']
})
df
```

```python
df.Name.str.split(' ',expand=True)
```

```python
df[['First Name','Last Name']] = df.Name.str.split(' ',expand=True)
df
```

```python
df.City.str.split('  ',expand=True)
```

```python
# Split 'City' column into 'City' and 'Country' columns
df[['City', 'Country']] = df['City'].str.split(' ', expand=True)
```

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    'Name': ['Asiif Khan', 'Naveed Ahmed', 'Kashif Khan', 'Asiif Khan', 'Muhammad Aammar', 'Whaeed Ahmad'],
    'Age': ['20', '21', '20', '20', '21', '20'],
    'City': ['Karachi pakistan', 'Multan pakistan', 'Fasilbad pakistan', 'Bahawalpur pakistan', 'Larkana pakistan', 'Mari pakistan']
})

# Display DataFrame
print("Original DataFrame:")
print(df)

# Split 'City' column into 'City' and 'Country' columns
df[['City', 'Country']] = df['City'].str.split(' ', n=1, expand=True)

# Display modified DataFrame
print("\nDataFrame after splitting 'City' column:")
print(df)
```

```python
df = df[["First Name","Last Name","Age","City","Country"]]
df
```

# 19-Reshape multi index series

```python
df.survived.mean()
```

```python
df.groupby('sex').survived.mean()
```

```python
df.groupby(['sex','class']).survived.mean()
```

The unstack function in Pandas is used to reshape a DataFrame with a hierarchical index by moving one level of the index from the rows to the columns.

unstack is useful for rearranging DataFrames with hierarchical indexing.\
It moves a level of the index from rows to columns.\
You can control the unstacking level and handle missing values with optional arguments.

```python
df.groupby(['sex','class']).survived.mean().unstack()
```

# 23- Reshaping a dataframe

```python
fasla = pd.DataFrame([[100,200,300],[400,500,600],[700,800,900]
                     ],columns=['a','b','c'])
fasla
```

```python
sns.barplot(data=fasla,x='b',y='a')
```

Putting it all together, the fasla.melt(id_vars=['a'], var_name='val', value_name='va') operation will transform the DataFrame fasla from wide format to long format, keeping the column 'a' as the identifier variable, and creating two new columns named 'val' and 'va' to store the variable names and values respectively.

```python
fasla.melt(id_vars=['a'], var_name='val', value_name='va')
```
