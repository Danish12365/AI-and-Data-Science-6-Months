# check that how maney MBs are there in one Kb? and how to calculate?
1. convert KB to MB
2. Then remove the M from all numbers\
3.  Handel`Varies with device`

```python
# Function to convert KB to MB
def convert_kb_to_mb(size):
    return size / 1024

# Function to remove 'M' from numbers
def remove_m_from_number(size):
    return size.rstrip('M')

# Convert 'Size' from KB to MB and remove 'M' from numbers
df['Size'] = df['Size'].apply(lambda x: convert_kb_to_mb(float(x.rstrip('M'))) if isinstance(x, str) and 'M' in x else x)

# Remove 'Varies with device' values
df['Size'] = df['Size'].apply(lambda x: remove_m_from_number(x) if isinstance(x, str) else x)
```

# do it as numaric column or group that into different group following bining method

```python
# Remove the '+' and ',' from 'Installs' column
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = df['Installs'].str.replace('+', '')

# Convert 'Installs' column to numeric
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Define the bin edges for grouping
bin_edges = [0, 1000, 100000, 10000000, np.inf]
bin_labels = ['0-1K', '1K-100K', '100K-10M', '10M+']

# Group 'Installs' into different bins
df['Installs Group'] = pd.cut(df['Installs'], bins=bin_edges, labels=bin_labels, right=False)
```
