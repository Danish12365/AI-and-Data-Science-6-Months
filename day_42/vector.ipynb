### Vectors

A vector is a mathematical object that has both magnitude (length) and direction. Vectors are often represented as arrows in a coordinate system, but they can also be represented algebraically as an ordered list of numbers, which are called the components of the vector. Vectors are fundamental in physics, engineering, and computer science.

#### Examples of Vectors:
1. **Displacement Vector**: Representing the position change from point A to point B.
   \[
   \mathbf{v} = \begin{pmatrix} 3 \\ 4 \end{pmatrix}
   \]
   This vector represents a displacement of 3 units in the x-direction and 4 units in the y-direction.

2. **Velocity Vector**: Representing the speed and direction of a moving object.
   \[
   \mathbf{v} = \begin{pmatrix} -2 \\ 1 \end{pmatrix}
   \]
   This vector represents a velocity of 2 units per second in the negative x-direction and 1 unit per second in the positive y-direction.

3. **Force Vector**: Representing a force applied in a particular direction.
   \[
   \mathbf{F} = \begin{pmatrix} 5 \\ -3 \\ 2 \end{pmatrix}
   \]
   This vector represents a force of 5 units in the x-direction, -3 units in the y-direction, and 2 units in the z-direction.

4. **Electric Field Vector**: Representing the electric field at a point in space.
   \[
   \mathbf{E} = \begin{pmatrix} 0 \\ 2 \\ 5 \end{pmatrix}
   \]
   This vector represents an electric field of 0 units in the x-direction, 2 units in the y-direction, and 5 units in the z-direction.

5. **Gradient Vector**: Representing the gradient of a scalar field.
   \[
   \nabla f = \begin{pmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{pmatrix}
   \]
   For a scalar function \(f(x, y) = x^2 + y^2\), the gradient vector is:
   \[
   \nabla f = \begin{pmatrix} 2x \\ 2y \end{pmatrix}
   \]

### Vector Spaces

A vector space (or linear space) is a collection of vectors that can be added together and multiplied ("scaled") by numbers, called scalars in this context. Scalars are often real numbers, but they can also be complex numbers or elements of other fields. A vector space must satisfy specific axioms related to vector addition and scalar multiplication.

#### Examples of Vector Spaces:
1. **Euclidean Space**: The set of all \(n\)-tuples of real numbers, often denoted \(\mathbb{R}^n\).
   \[
   \mathbb{R}^3 = \left\{ \begin{pmatrix} x \\ y \\ z \end{pmatrix} \mid x, y, z \in \mathbb{R} \right\}
   \]
   This is a 3-dimensional vector space over the real numbers.

2. **Polynomial Space**: The set of all polynomials of a given degree or less.
   \[
   P_2 = \{a + bx + cx^2 \mid a, b, c \in \mathbb{R} \}
   \]
   This is a vector space of polynomials of degree 2 or less.

3. **Matrix Space**: The set of all \(m \times n\) matrices.
   \[
   M_{2x2} = \left\{ \begin{pmatrix} a & b \\ c & d \end{pmatrix} \mid a, b, c, d \in \mathbb{R} \right\}
   \]
   This is the vector space of all \(2 \times 2\) matrices.

4. **Function Space**: The set of all continuous functions on a given interval.
   \[
   C[0, 1] = \{ f \mid f \text{ is continuous on } [0, 1] \}
   \]
   This is the vector space of all continuous functions on the interval \([0, 1]\).

5. **Solution Space of a Linear System**: The set of all solutions to a given homogeneous linear differential equation.
   \[
   y'' + y = 0
   \]
   The solutions form a vector space because any linear combination of solutions is also a solution. For this particular equation, the general solution is:
   \[
   y = C_1 \cos(x) + C_2 \sin(x)
   \]
   where \(C_1\) and \(C_2\) are constants.

These examples illustrate the diverse contexts in which vectors and vector spaces appear, spanning from geometry and physics to more abstract mathematical settings.
