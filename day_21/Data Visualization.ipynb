```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
df1 =sns.load_dataset('iris')
```

```python
df1.columns
```

# 1. Line Plot:

```python
# Create a line plot
plt.figure(figsize=(8, 6))  # Adjust figure size (optional)
df1.plot(kind='line', x='sepal_length', y='sepal_width', marker='o', color='b')  # Customize marker and color
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('petal_length and sepal_width')
plt.grid(True)  # Add gridlines (optional)
plt.show()
```

# 2. Scatter Plot:

```python
plt.figure(figsize=(8, 6))
df1.plot(kind='scatter', x='petal_length', y='petal_width', color='g')  # Customize color
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('petal_length vs. petal_width')
plt.show()
```

# 3. Histogram (Petal Length Distribution):

```python
plt.figure(figsize=(8, 6))
df1.plot(kind='hist', x='petal_length', y='petal_width', color='g')  # Customize color
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length")
plt.show()
```

```python
plt.figure()
df1['sepal_width'].plot(kind="bar")
```

# 4. KDE Plot (Petal Width Kernel Density Estimation):

```python
plt.figure(figsize=(8, 6))
df1.plot(kind='hist', x='petal_width', y='Density', color='g')  # Customize color
plt.xlabel("petal_width")
plt.ylabel("Density")
plt.title("Kernel Density Estimation of Petal Width")
plt.show()
```
