# 1- USING PANDAS FOR CREATING PLOTS

```python
df1 =sns.load_dataset('iris')
```

```python
df1.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot 1: Histogram of sepal_length
df1['sepal_length'].hist()
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()

# Plot 2: Histogram of sepal_width
df1['sepal_width'].hist()
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.show()

# Plot 3: Scatter plot of sepal_length vs sepal_width with different colors for species
plt.scatter(x=df1['sepal_length'], y=df1['sepal_width'], c=df1['species'].astype('category').cat.codes, cmap='viridis')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.colorbar(label='Species')
plt.show()


# Plot 4: Box plot of petal_length by species
df1.boxplot(column='petal_length', by='species')
plt.title('Box Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()

# Plot 5: Line plot of petal_length
df1['petal_length'].plot()
plt.title('Line Plot of Petal Length')
plt.xlabel('Index')
plt.ylabel('Petal Length')
plt.show()

# Plot 6: Bar plot of mean petal_width by species
df1.groupby('species')['petal_width'].mean().plot(kind='bar')
plt.title('Mean Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Mean Petal Width')
plt.show()

# Plot 7: Box plot of all numeric columns
df1.boxplot()
plt.title('Box Plot of All Numeric Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()

# Plot 8: Density plot of petal_length
df1['petal_length'].plot.density()
plt.title('Density Plot of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Density')
plt.show()

# Plot 9: Bar plot of species counts
df1['species'].value_counts().plot(kind='bar')
plt.title('Bar Plot of Species Counts')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

# Plot 10: Area plot of petal_width
df1['petal_width'].plot.area()
plt.title('Area Plot of Petal Width')
plt.xlabel('Index')
plt.ylabel('Petal Width')
plt.show()
```

# 3- Creating plots using Matplotlib

```python
df3 = sns.load_dataset('tips')
```

```python
df3.head()
```

# 1- Bar Plot:
Plot the total_bill for each day:

```python
plt.bar(df3['day'], df3['total_bill'])
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')
plt.show()
```

# 2-Histogram:
Plot the distribution of total_bill:

```python
plt.hist(df3['total_bill'], bins=10)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

# 3-Scatter Plot:
Plot tip vs. total_bill:

```python
plt.scatter(df3['total_bill'], df3['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Tip vs. Total Bill')
plt.show()
```

# 4-Box Plot:
Plot the distribution of total_bill for each day:

```python
plt.boxplot([df3[df3['day'] == day]['total_bill'] for day in df3['day'].unique()], labels=df3['day'].unique())
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill Distribution by Day')
plt.show()
```

# 5-Pie Chart:
Plot the proportion of male and female customers:

```python
plt.pie(df3['sex'].value_counts(), labels=df3['sex'].unique(), autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

# 6-Line Plot:
Plot the trend of total_bill over time:

```python
plt.plot(df3.index, df3['total_bill'])
plt.xlabel('Index')
plt.ylabel('Total Bill')
plt.title('Total Bill Trend over Time')
plt.show()
```

# 7-Stacked Bar Plot:
Plot the total_bill and tip for each day as a stacked bar plot:

```python
plt.bar(df3['day'], df3['total_bill'], label='Total Bill')
plt.bar(df3['day'], df3['tip'], bottom=df3['total_bill'], label='Tip')
plt.xlabel('Day')
plt.ylabel('Amount')
plt.title('Total Bill and Tip by Day')
plt.legend()
plt.show()
```

# 8-Violin Plot:
Plot the distribution of total_bill for each day using a violin plot:

```python
plt.violinplot([df3[df3['day'] == day]['total_bill'] for day in df3['day'].unique()], showmeans=True)
plt.xticks(range(1, len(df3['day'].unique()) + 1), df3['day'].unique())
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill Distribution by Day')
plt.show()
```

# 9-Hexbin Plot:
Plot a hexbin plot of total_bill vs. tip:

```python
plt.hexbin(df3['total_bill'], df3['tip'], gridsize=20)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Hexbin Plot of Total Bill vs. Tip')
plt.colorbar(label='count')
plt.show()
```

# 10-Stacked Area Plot:
Plot the cumulative sum of total_bill over time as a stacked area plot:

```python
plt.stackplot(df3.index, df3['total_bill'].cumsum(), labels=['Total Bill'])
plt.xlabel('Index')
plt.ylabel('Cumulative Total Bill')
plt.title('Cumulative Total Bill over Time')
plt.legend()
plt.show()
```

# 4-Creating plots using Plotly 

```python
data = sns.load_dataset('exercise')
```

```python
import plotly.graph_objects as go
import plotly.express as px

# Scatter Plot
scatter_fig = px.scatter(data_frame=data, x='pulse', y='time', color='diet', title='Scatter Plot of Pulse vs. Time by Diet')
scatter_fig.show()

# Line Plot
line_fig = px.line(data_frame=data, x='Unnamed: 0', y='pulse', color='kind', title='Line Plot of Pulse over Time')
line_fig.show()

# Bar Plot
bar_fig = px.bar(data_frame=data, x='diet', y='pulse', title='Bar Plot of Pulse by Diet')
bar_fig.show()

# Box Plot
box_fig = px.box(data_frame=data, x='kind', y='pulse', color='diet', title='Box Plot of Pulse by Exercise Kind')
box_fig.show()

# Histogram
hist_fig = px.histogram(data_frame=data, x='pulse', title='Histogram of Pulse')
hist_fig.show()

# Violin Plot
violin_fig = px.violin(data_frame=data, x='diet', y='pulse', box=True, title='Violin Plot of Pulse by Diet')
violin_fig.show()

# Pie Chart
pie_fig = px.pie(data_frame=data, names='diet', title='Pie Chart of Diet Distribution')
pie_fig.show()

# Area Plot
area_fig = px.area(data_frame=data, x='Unnamed: 0', y='pulse', title='Area Plot of Pulse over Time')
area_fig.show()

# Heatmap
heatmap_fig = px.density_heatmap(data_frame=data, x='diet', y='pulse', title='Heatmap of Pulse by Diet')
heatmap_fig.show()

# Scatter Matrix Plot
scatter_matrix_fig = px.scatter_matrix(data_frame=data, dimensions=['pulse', 'Unnamed: 0'], color='kind',
                                       title='Scatter Matrix Plot of Pulse and Time')
scatter_matrix_fig.show()

# Line Plot with Error Bars
line_error_fig = px.line(data_frame=data, x='time', y='pulse', error_y='pulse', title='Line Plot with Error Bars')
line_error_fig.show()

# 3D Scatter Plot
scatter_3d_fig = px.scatter_3d(data_frame=data, x='pulse', y='time', z='Unnamed: 0', color='kind',
                                title='3D Scatter Plot of Pulse, Time, and Index')
scatter_3d_fig.show()

# Contour Plot
contour_fig = px.density_contour(data_frame=data, x='pulse', y='time', title='Contour Plot of Pulse and Time')
contour_fig.show()

# Polar Plot
polar_fig = px.scatter_polar(data_frame=data, r='pulse', theta='time', color='diet', title='Polar Plot of Pulse and Time by Diet')
polar_fig.show()

```
