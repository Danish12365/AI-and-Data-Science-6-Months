# Cartesian Coordinates

Cartesian coordinates are a system used to uniquely determine the position of a point in a plane or space by using two or three numerical values, respectively. These values represent the distances from the point to two or three mutually perpendicular lines called axes.

### Examples of Cartesian Coordinates

#### 1. Points in a 2D Plane
In a 2D Cartesian coordinate system, each point is determined by an ordered pair \((x, y)\).

1. **Point A**
   \[
   A = (3, 4)
   \]

2. **Point B**
   \[
   B = (-2, 1)
   \]

3. **Point C**
   \[
   C = (0, -3)
   \]

4. **Point D**
   \[
   D = (5, 5)
   \]

5. **Point E**
   \[
   E = (-3, -4)
   \]

#### 2. Points in a 3D Space
In a 3D Cartesian coordinate system, each point is determined by an ordered triple \((x, y, z)\).

1. **Point F**
   \[
   F = (2, 3, 5)
   \]

2. **Point G**
   \[
   G = (-1, 4, -2)
   \]

3. **Point H**
   \[
   H = (0, -3, 3)
   \]

4. **Point I**
   \[
   I = (4, 0, 2)
   \]

5. **Point J**
   \[
   J = (-2, -1, -1)
   \]
