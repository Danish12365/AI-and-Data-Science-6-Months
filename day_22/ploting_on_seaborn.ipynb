```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
palette  = {'male' : 'green','female' : 'red'}
plt.figure(figsize=(10,5))    # A4 for 10,8 and 16:9 for slides
sns.scatterplot(data = df, x = 'age', y = 'fare', hue= 'sex', palette = palette )     # scatter plot is use on numeric data 
plt.xlabel('Age' , fontsize=15)
plt.ylabel('Fare' , fontsize=15)
plt.title('Age vs Fare' , fontsize=20)
plt.show()


# saving a plot in 300dpi, 600dpi, 150dpi png 
plt.savefig('./outputs/titanic.png', dpi=150)

# saving a plot in vector file such as svg
plt.savefig('./outputs/titanic.svg', dpi=150, format='svg')

```

```python
# style of plot
sns.set_theme(style="darkgrid", font_scale=1.5,font='times new roman')
palette  = {'male' : 'green','female' : 'red'}
plt.figure(figsize=(10,5))    # A4 for 10,8 and 16:9 for slides
sns.scatterplot(data = df, x = 'age', y = 'fare', hue= 'sex', palette = palette )
plt.xlabel('Age' , fontsize=15)
plt.ylabel('Fare' , fontsize=15)
plt.title('Age vs Fare' , fontsize=20)
# change legend position and title
plt.legend(loc='upper right', title='Sex')
plt.show()
```

```python
# style of plot
sns.set_theme(style="darkgrid", font_scale=1.5,font='times new roman')
palette  = {'male' : 'green','female' : 'red'}
plt.figure(figsize=(10,5))    # A4 for 10,8 and 16:9 for slides
sns.scatterplot(data = df, x = 'age', y = 'fare', hue= 'sex', size='fare', palette = palette ) # we can change the size of the dots of the plot
plt.xlabel('Age' , fontsize=15)
plt.ylabel('Fare' , fontsize=15)
plt.title('Age vs Fare' , fontsize=20)
# change legend position and title
plt.legend(loc='upper right', title='Sex')
plt.show()
```

```python
sns.set_theme(style="darkgrid", font_scale=1.5,font='times new roman')
palette  = {'First' : 'green','Second' : 'red', 'Third' : 'blue'}
plt.figure(figsize=(10,5))    # A4 for 10,8 and 16:9 for slides
sns.boxplot(data = df, x = 'sex', y = 'age', hue= 'class', palette = palette )
plt.xlabel('Gender' , fontsize=15)
plt.ylabel('AGE' , fontsize=15)
plt.title('Box Plot vs Fare' , fontsize=20)
# change legend position and title
plt.legend(loc='upper right', title='Sex')
plt.show()
```

```python
sns.set_theme(style="whitegrid", font_scale=1.5,font='times new roman')
palette  = {'First' : 'green','Second' : 'red', 'Third' : 'blue'}
plt.figure(figsize=(10,5))    # A4 for 10,8 and 16:9 for slides
sns.boxenplot(data = df, x = 'sex', y = 'age', hue= 'class', palette = palette )
plt.xlabel('Gender' , fontsize=15)
plt.ylabel('AGE' , fontsize=15)
plt.title('Box Plot vs Fare' , fontsize=20)
# change legend position and title
plt.legend(loc='upper right', title='Sex')
# change the x-tick names
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=45)
plt.show()
```

```python
sns.set_theme(style="whitegrid", font_scale=1.5,font='times new roman')
palette  = {'First' : 'green','Second' : 'red', 'Third' : 'blue'}
plt.figure(figsize=(10,5))    # A4 for 10,8 and 16:9 for slides
sns.barplot(data = df, x = 'sex', y = 'age', hue= 'class', palette = palette, capsize=0.1 ) # cap size of error bar
plt.xlabel('Gender' , fontsize=15)
plt.ylabel('AGE' , fontsize=15)
plt.title('Bar Plot vs Fare' , fontsize=20)
# change legend position and title
plt.legend(loc='upper right', title='Sex')
# change the x-tick names
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=45)
# add caps to the error bars
plt.errorbar(x=[0,1], y=[30, 28], yerr=[2,3], fmt=' ', color='black', capsize=5)
plt.show()
```
